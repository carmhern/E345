```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import pandas
from scipy import signal

#from google.colab import drive

```

```python
s1 = pandas.read_csv("sustain_test1.csv")
s2 = pandas.read_csv("sustain_test2.csv")
s1 = s1[1:]


p1 = pandas.read_csv("protocall1.csv")
p2 = pandas.read_csv("protocall2.csv")

```

```python
#estimating sustained grip 1

#beginning of grip

bs1 = s1[0:900]

ft_bs1 = np.fft.rfft(bs1)
ft_bs1 = np.abs(ft_bs1)
ps_bs1 = np.square(ft_bs1) / (900**2)

plt.plot(ps_bs1)
plt.title('Power Specteral Density of the Beginning of Sustained Grip 1')
plt.xlabel('Samples')
plt.ylabel('Voltage')
plt.show()

#middle of grip
ms1 = s1[901:1800]

ft_ms1 = np.fft.rfft(ms1)
ft_ms1 = np.abs(ft_ms1)
ps_ms1 = np.square(ft_ms1) / (900**2)

plt.plot(ps_ms1)
plt.xlabel('Samples')
plt.ylabel('Voltage')
plt.xticks([0, 200, 400, 600, 800], [900, 1100, 1300, 1500, 1700])
plt.title('Power Specteral Density of the Middle of Sustained Grip 1')
plt.show()

#end of grip
es1 = s1[1801:2754]

ft_es1 = np.fft.rfft(es1)
ft_es1 = np.abs(ft_es1)
ps_es1 = np.square(ft_es1) / (953**2)

plt.plot(ps_es1)
plt.xlabel('Samples')
plt.ylabel('Voltage')
plt.xticks([0, 200, 400, 600, 800, 1000], [1800, 2000, 2200, 2400, 2600])
plt.title('Power Specteral Density of the End of Sustained Grip 1')
plt.show()
```

The power specteral density of beginning, middle and end of the sustained test are shown in the graphs above.  The beginning portion of the test shows us before the muscles were activated, a huge spike from the initial use of the muscle, and then a dip to the voltage that is able to be sustained.   The graph then has small dips and spikes from the muscle struggling to maintain one constant voltage.  At the end of the test we see the muscle begin to relax and finally a huge drop before it stops. 

After visually inspecting the EMG signals for our protocol there is nothing that is out of the ordinary for our tests.  We are able to see each grip and the increase in power to try and reach the desired goal.  No artifacts or disrupts are seen in either test.  The 2nd test is not as clean as the first. I was the test subject for this and on the 2nd test I was keeping my forearm slightly engaged to have more control.  This would explain why the rest periods are not as clean.
