```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import scipy


#plot empty fs1kHz.csv file

File = np.loadtxt('empty_fs1kHz.csv')

plt.xlabel('Time [sec]')
plt.ylabel('Voltage')
plt.title('Time Plot of Signal')
plt.plot(File)
plt.xticks([0, 2000, 4000, 6000, 8000, 10000], [0, 2, 4, 6, 8, 10])
plt.show()
```

```python
#plot 25 seconds of signal

x = np.arange(0,26)
plt.xlabel('Time [sec]')
plt.ylabel('Voltage')
plt.title('Time Plot of Signal for 2.5 Seconds')
plt.plot(File[0:2500])
plt.xticks([0, 500, 1000, 1500, 2000, 2500], [0, .5, 1, 1.5, 2, 2.5])
plt.show()
```

```python

#plot specteral anaylsis of full signal

plt.magnitude_spectrum(File)
plt.title('Magnitude Plot of Signal')
plt.show()

plt.phase_spectrum(File)
plt.title('Phase Plot of Signal')
plt.ylabel('Phase')
plt.xlabel('Frequency [Hz]')
plt.show()
```

```python
#plot signal with sampling rate of 100 Hz

file1 = scipy.signal.resample(File, 100)
plt.plot(file1)
plt.title('Time Plot of 100Hz Sampling Rate Signal')
plt.xlabel('Time [sec]')
plt.xticks([0, 20, 40, 60, 80, 100], [0, 2, 4, 6, 8, 10])
plt.ylabel('Voltage')
plt.show()


plt.magnitude_spectrum(file1)
plt.title('Magnitude Plot of 100Hz Sampling Rate Signal')
plt.show()
```

```python
#plot with sampling rate of 50 Hz

file2 = scipy.signal.resample(File, 50)
plt.plot(file2)
plt.title('Time Plot of 50Hz Sampling Rate Signal')
plt.xlabel('Time [sec]')
plt.xticks([0, 10, 20, 30, 40, 50], [0, 2, 4, 6, 8, 10])
plt.ylabel('Voltage')
plt.show()

plt.magnitude_spectrum(file2)
plt.title('Magnitude Plot of 50Hz Sampling Rate Signal')
plt.show()

```

```python
#plot High Pass filter

HP = np.loadtxt('HP_fs100Hz_fc40Hz.csv')
LP = np.loadtxt('LP_fs100Hz_fc40Hz.csv')

convH = np.convolve(HP, file1)

plt.plot(convH)
plt.title('Time Plot of Convolution of High Pass Filter')
plt.xlabel('Time [sec]')
plt.ylabel('Voltage')
plt.show()

plt.magnitude_spectrum(convH)
plt.title('Magnitude of HP Filter')
plt.show()
```

```python
#plot Low Pass filter

convL = np.convolve(LP, file1)

plt.plot(convL)
plt.title('Time Plot of Convolution of Low Pass Filter')
plt.xlabel('Time [sec]')
plt.ylabel('Voltage')
plt.show()

plt.magnitude_spectrum(convL)
plt.title('Magnitude Plot of LP Filter')
plt.show()
```

```python
#plot Low Pass Filter at 50 Hz

final = scipy.signal.resample(convL, 50)

plt.plot(final)
plt.title('Time Plot of LP Filter at 50 Hz')
plt.xlabel('Time [sec]')
plt.ylabel('Voltage')
plt.show()

plt.magnitude_spectrum(final)
plt.title('Magnitude Plot of LP Filter at 50 Hz')
plt.show()
```
